1. SWOT stands for Strengths, Weaknesses, Oppurtunities and Threats. SWOT Analysis is a strategic tool to help identify certain objectives.

To perform a SWOT Analysis on the Google Merchandise Store Website, the steps are:

- Defining the objectives that we aim to achieve with the help of SWOT Analysis.

- Identifying strengths such as unique selling proposition, customer satisfaction, great UI interface of the website.

- Identifying weaknesses such as low conversion rate, high bounce rate, website loading speed, checkout processes.

- Identifying oppurtunities such as new marketting channels, new products, new UI interface, faster checkout processes.

- Identifying threats such as similar market competition, change in customer's preferences.

- Based on the above SWOT Analysis, an action plan needs to be framed keeping in mind the strength, weakness, threats and oppurtunities.

- After making some changes, we need to reviee the steps taken and if need arises make necessary changes and monitor again.

2. Features of the dataset on which we can perform A/B Testing are:

- UI Interface

- Product Pricing

- Product Images

- Navigation Menus

- Checkout Pages

3. The different types of A/B Tests that can be performed on the datatsets are:

- Funnel Test - Test the conversion rate of a step by step process like a checkout or new registration/sign up form.

- Split Test - Testing two different versions of web pages with different URLs to check which webpages perform better.

- Price Test - Testing pricing strategies like discounts, bundling different products to check with strategy performs better.

- ReDirect Test - Testing different versions of a web page by redirecting visitors to different pages.

- Personalise Test - Testing different variations of a web page based on different locations and also different profiles of customer.

- Landing Page Optimisation

- Online Experimentation

4. The statistical approach that can be used to to run an A/B Test is Hypothesis Testing.

5. Steps to perform an A/B Tests are:

- Selecting a Testing Tool that are mentioned above such as Funnel Tests or Split Test, etc.

- Defining Primary Success Metric like Bounce Rate, Exit Rate, etc.

- Defining Idea/Questions/Hyptothesis that needs to be tested.

- Defining Variations and Learning

- Creating the variations and structure.

- Measure and Activate - Activate for an experiment basis for a shorter time and analyse the improvement.

- Analysing the results.

- Document the learning.

- If the results are deemed to be success, the above variation is deployed. If not, above steps are repeated.

6. Advantages of A/B Testing for this website are:

- Increase Revenue and Conversions.

- Uses actual site visitors.

- Data-Driven Decision Making.

- Reduces risk of making mistakes.

- Provides data driven insgihts to make informed decisions.

- Rapid Iteration.

7. The challenges that we can face while performing A/B testing are:

- Not able to select right metric can lead to less effectiveness of test.

- Not enough sample size.

- Not learning from failed tests.

- Testing multiple elements at the same time.

- Influence of external factors.
